Dataset Overview

This dataset contains information on beauty and fragrance products, including product
names, brands, customer engagement metrics such as loves count, ratings, and the number
of reviews.


```python
import pandas as pd
import matplotlib.pyplot as plt


products = pd.read_csv("../data/product_info.csv")
brands = pd.read_csv("../data/brand_info.csv")

merged = pd.merge(products, brands, on="brand_name", how="left")

merged.head()


```

```python
merged.shape
merged.columns
merged.head()
```

```python
merged.info()
merged.describe()
```

```python
merged.isna().sum().sort_values(ascending=False)
```

What the data looks like
- Each row is one product/variation.
- Columns include product + brand details and customer engagement metrics (loves, rating, reviews).
- Some fields (like size/variation details) are missing for certain rows, which probably means
  those products don’t have that info available or don’t have variations listed.


```python
merged["brand_name"].value_counts().head(10)
```

```python
merged["rating"].mean()
```

Summary Observation
summary statistics:
- The majority of product ratings are above average and products are positively rated.
- Some brands are represented more than others, meaning they have more products in the dataset.
- The number of reviews and loves varies wildly, suggesting that some products are far more popular than others on the platform.

```python
merged["rating"].plot(kind="hist", title="Distribution of Product Ratings")
```

```python
merged["brand_name"].value_counts().head(10).plot(
    kind="bar",
    title="Top 10 Brands by Number of Products"
)
```

```python
merged.groupby("brand_name")["rating"].mean().sort_values(ascending=False).head(10).plot(
    kind="bar",
    title="Top 10 Brands by Average Rating"
)
```

Graph Overview

graph 1 shows the distribution of product ratings across the dataset. Most products are rated between 4.0 and 5.0, indicating that customer feedback is generally positive. Very few products have low ratings, suggesting that poorly rated products are less common.

graph 2 this bar chart shows the top 10 brands based on the number of products
they have in the dataset. A small number of brands account for a large portion of the products,
which suggests that the retailer carries more items from certain
wellknown or popular brands.

graph 3 This chart displays the top 10 brands ranked by their average product rating. Brands with higher average ratings tend to have more consistently well reviewed products. This suggests that customer satisfaction may vary by brand, not just by individual product.




Summary Statistics

Basic summary statistics were used to understand the overall structure
of the dataset. Ratings are generally high across products, while the number of reviews
and loves varies widely. This shows that some products receive much more attention and engagement from customers than others.



Insights

Overall, the dataset shows strong customer engagement with beauty
products. Most products receive positive ratings, and certain brands
stand out in terms of both product count and customer satisfaction. This helps find which brands and products perform well and could be useful for decisions related to marketing or product focus.

## Analysis Using Clean Joined Dataset


```python
import pandas as pd

clean = pd.read_csv("../clean_data/products_with_brand.csv")

clean.shape
clean.head()
```

### Summary Statistics


```python
clean.describe()
```

We calculate the average rating by brand to see which brands perform best.


```python
clean.groupby("brand_name")["rating"].mean().sort_values(ascending=False).head(10)
```

Products with the highest number of loves are the most popular.

```python
clean.sort_values("loves_count", ascending=False)[["product_name","brand_name","loves_count"]].head(10)
```

We compare reviews and rating to see if popular products are rated higher.


```python
top_brands = (
    clean.groupby("brand_name")["rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_brands.plot(kind="bar", title="Top 10 Brands by Average Rating", figsize=(10,5))
plt.xticks(rotation=45)
plt.tight_layout()
```

Most Loved Products

```python
top_products = clean.sort_values("loves_count", ascending=False).head(10)

top_products.plot(
    x="product_name",
    y="loves_count",
    kind="bar",
    title="Top 10 Most Loved Products",
    figsize=(10,5)
)

plt.xticks(rotation=70)
plt.tight_layout()
```
